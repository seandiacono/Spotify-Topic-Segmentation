```python
from load_transcripts import load_transcripts
```

```python
transcripts = load_transcripts(path='../spotify-podcasts-2020/podcasts-transcripts/', limit=100)
```

```python
from nltk.tokenize.texttiling import TextTilingTokenizer
```

```python
tt = TextTilingTokenizer()
text = tt.tokenize(transcripts[0])
```

```python
for topic in text:
    print('New topic:')
    print(topic)
```

```python
# Save text to file
with open('transcripts_tiling.txt', 'w') as f:
    for topic in text:
        f.write(topic)
        f.write('\n==========\n')
```
